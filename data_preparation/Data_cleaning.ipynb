```python
import pandas as pd
```

```python
reviews = pd.read_csv('hotels_reviews.csv')
```

```python
reviews.shape
```

```python
reviews
```

```python
reviews.info()
```

```python
# rimozione duplicati
reviews = reviews.drop_duplicates()
reviews
```

```python
#drop colonne non utili
reviews = reviews.drop(columns = ['country', 'latitude', 'longitude','reviews.dateAdded', 'reviews.doRecommend', 'reviews.id', 'reviews.userCity', 'reviews.userProvince'])
reviews
```

```python
#valori mancanti
reviews.isnull().sum()
```

```python
#drop reviews con valori mancanti per gli attributi visti sopra
reviews = reviews.dropna(subset = ['postalCode', 
                                   'reviews.date', 
                                   'reviews.rating', 
                                   'reviews.text', 
                                   'reviews.title', 
                                   'reviews.username'])
```

```python
reviews.isnull().sum()
```

```python
reviews['reviews.rating'].value_counts()
```

```python
#formattare reviews date mantenendo solo la data, senza il tempo
reviews['reviews.date'] = reviews['reviews.date'].str.split('T').str[0]
reviews['reviews.date']
```

```python
#filtraggio delle sole reviews con ratings in [1 ,2, 3, 4, 5]
ratings_corretti = [1, 2, 3, 4 ,5]
reviews = reviews[reviews['reviews.rating'].isin(ratings_corretti)]
reviews['reviews.rating'].value_counts()
```

```python
reviews
```

```python
#drop reviews con carattere � nel titolo
reviews = reviews[~reviews['reviews.title'].str.contains('�', na=False)]
reviews
```

```python
#save into new csv
reviews.to_csv('cleaned_hotels_reviews.csv', index = False)
```
